# Plot Tracks

```python
import matplotlib.pyplot as plt
from cartopy.crs import PlateCarree, EqualEarth

from storm_assess import SAMPLE_TRACK_DATA, track, plot
```

```python
storms = track.load_no_assumptions(SAMPLE_TRACK_DATA)
```

```python
transform = PlateCarree()
projection = EqualEarth()

ax = plt.axes(projection=projection)
for n in range(10):
    plot.plot_track(storms[n], xname="longitude", yname="latitude", transform=transform)
ax.coastlines()
ax.gridlines()
```
